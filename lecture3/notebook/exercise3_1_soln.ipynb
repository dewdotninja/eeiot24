```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl
```

```python
Rs = 10000
Rf = 100000
fo = 2 # Hz
wo = 2*np.pi*fo
Cf = 1/(wo*Rf)
Cf
```

```python
fb = 1
wb = 2*np.pi*fb
fn = 50
wn = 2*np.pi*fn
tvec = np.arange(0,3,0.001)
xb = np.sin(wb*tvec)
xn = 0.2*np.sin(wn*tvec)
x = xb+xn
plt.figure()
plt.plot(tvec,x,'r-')
```

```python
s = ctl.tf("s")
H = -(Rf/Rs)/(Cf*Rf*s+1)
H
```

```python
_,y = ctl.forced_response(H,tvec,x)
```

```python
plt.plot(tvec,y,'b-')
```
